```python
""" 
–†—è–¥ –§–∏–±–æ–Ω–∞—á—á–∏ ‚Äî –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π –ø—Ä–∏–º–µ—Ä —Ä–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.
–†–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω–æ—Å—Ç—å: –õ—é–±–æ–µ —á–∏—Å–ª–æ, –∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ –∏ –≤—Ç–æ—Ä–æ–≥–æ, –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∫–∞–∫ —Å—É–º–º–∞ –¥–≤—É—Ö –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö.

–û—Å–Ω–æ–≤–Ω–∞—è —Ñ–æ—Ä–º—É–ª–∞ fib(n) = fib(n-1) + fib(n-2) –ø—Ä–∏ fib(1)=0, fib(2)=1.

–ò—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –º–µ—Ç–æ–¥ –ø—Ä–µ–¥–ø–æ—á—Ç–∏—Ç–µ–ª–µ–Ω –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.


–ß—Ç–æ —Ç–∞–∫–æ–µ —Ä—è–¥ –§–∏–±–æ–Ω–∞—á—á–∏? –ü—Ä–æ—Å—Ç—ã–º —è–∑—ã–∫–æ–º
–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ, —á—Ç–æ –≤—ã –∫—Ä–æ–ª–∏–∫–æ–≤–æ–¥:

–ú–µ—Å—è—Ü 1: –£ –≤–∞—Å –µ—Å—Ç—å 1 –ø–∞—Ä–∞ –Ω–æ–≤–æ—Ä–æ–∂–¥–µ–Ω–Ω—ã—Ö –∫—Ä–æ–ª–∏–∫–æ–≤ (—Å–∞–º–µ—Ü –∏ —Å–∞–º–∫–∞)

–ú–µ—Å—è—Ü 2: –û–Ω–∏ –ø–æ–¥—Ä–∞—Å—Ç–∞—é—Ç, –Ω–æ –ø–æ—Ç–æ–º—Å—Ç–≤–∞ –Ω–µ—Ç ‚Üí 1 –ø–∞—Ä–∞

–ú–µ—Å—è—Ü 3: –û–Ω–∏ –¥–∞—é—Ç –ø–æ—Ç–æ–º—Å—Ç–≤–æ ‚Äî 1 –Ω–æ–≤–∞—è –ø–∞—Ä–∞ ‚Üí –≤—Å–µ–≥–æ 2 –ø–∞—Ä—ã

–ú–µ—Å—è—Ü 4: –ü–µ—Ä–≤—ã–µ –∫—Ä–æ–ª–∏–∫–∏ —Å–Ω–æ–≤–∞ –¥–∞—é—Ç –ø–æ—Ç–æ–º—Å—Ç–≤–æ ‚Üí +1 –ø–∞—Ä–∞, –≤—Å–µ–≥–æ 3 –ø–∞—Ä—ã

–ú–µ—Å—è—Ü 5: –î–≤–µ –≤–∑—Ä–æ—Å–ª—ã–µ –ø–∞—Ä—ã –¥–∞—é—Ç –ø–æ—Ç–æ–º—Å—Ç–≤–æ ‚Üí +2 –ø–∞—Ä—ã, –≤—Å–µ–≥–æ 5 –ø–∞—Ä

–ß–∏—Å–ª–∞ –ø–∞—Ä: 1, 1, 2, 3, 5, 8, 13, 21...
–≠—Ç–æ –∏ –µ—Å—Ç—å —Ä—è–¥ –§–∏–±–æ–Ω–∞—á—á–∏ ‚Äî –∫–∞–∂–¥–∞—è —Å–ª–µ–¥—É—é—â–∞—è –ø–∞—Ä–∞ —Ä–∞–≤–Ω–∞ —Å—É–º–º–µ –¥–≤—É—Ö –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö.

–ì–¥–µ —ç—Ç–æ –ù–£–ñ–ù–û –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏? (–ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ)
üìç 1. –§–∏–Ω–∞–Ω—Å—ã –∏ —Ç—Ä–µ–π–¥–∏–Ω–≥
–£—Ä–æ–≤–Ω–∏ –§–∏–±–æ–Ω–∞—á—á–∏ –≤ —Ç–µ—Ö–Ω–∏—á–µ—Å–∫–æ–º –∞–Ω–∞–ª–∏–∑–µ ‚Äî —Ç—Ä–µ–π–¥–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑—É—é—Ç —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è 23.6%, 38.2%, 61.8% –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É—Ä–æ–≤–Ω–µ–π –ø–æ–¥–¥–µ—Ä–∂–∫–∏/—Å–æ–ø—Ä–æ—Ç–∏–≤–ª–µ–Ω–∏—è

–†–µ—Ç—Ä–µ–π—Å–º–µ–Ω—Ç—ã –§–∏–±–æ–Ω–∞—á—á–∏ ‚Äî –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –∫–æ—Ä—Ä–µ–∫—Ü–∏–π –¥–≤–∏–∂–µ–Ω–∏—è —Ü–µ–Ω—ã

üìç 2. –ö–æ–º–ø—å—é—Ç–µ—Ä–Ω—ã–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã
–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —Ä–µ–∫—É—Ä—Å–∏–∏ ‚Äî –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–∏–π —É—á–µ–±–Ω—ã–π –ø—Ä–∏–º–µ—Ä

–ê–Ω–∞–ª–∏–∑ —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤ ‚Äî –æ—Ü–µ–Ω–∫–∞ –≤—Ä–µ–º–µ–Ω–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ä–µ–∫—É—Ä—Å–∏–≤–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π

–î–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–µ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ ‚Äî –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è –º–µ–º–æ–∏–∑–∞—Ü–∏–∏ (–∫—ç—à–∏—Ä–æ–≤–∞–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤)

üìç 3. –ö—Ä–∏–ø—Ç–æ–≥—Ä–∞—Ñ–∏—è
–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä—ã –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª –∏—Å–ø–æ–ª—å–∑—É—é—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –§–∏–±–æ–Ω–∞—á—á–∏

üìç 4. –ü—Ä–∏—Ä–æ–¥–∞ –∏ –¥–∏–∑–∞–π–Ω
–†–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ –ª–∏—Å—Ç—å–µ–≤ –Ω–∞ —Å—Ç–µ–±–ª–µ (—Ñ–∏–ª–ª–æ—Ç–∞–∫—Å–∏—Å) ‚Äî —É–≥–ª—ã –º–µ–∂–¥—É –ª–∏—Å—Ç—å—è–º–∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –∑–æ–ª–æ—Ç–æ–º—É —Å–µ—á–µ–Ω–∏—é

–°–ø–∏—Ä–∞–ª–∏ –≤ –ø–æ–¥—Å–æ–ª–Ω—É—Ö–µ, –∞–Ω–∞–Ω–∞—Å–µ, —Å–æ—Å–Ω–æ–≤–æ–π —à–∏—à–∫–µ ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ø–∏—Ä–∞–ª–µ–π –≤–ª–µ–≤–æ –∏ –≤–ø—Ä–∞–≤–æ —è–≤–ª—è—é—Ç—Å—è —á–∏—Å–ª–∞–º–∏ –§–∏–±–æ–Ω–∞—á—á–∏

–†–∞–∫–æ–≤–∏–Ω–∞ –Ω–∞—É—Ç–∏–ª—É—Å–∞ ‚Äî —Ä–∞—Å—Ç–µ—Ç –ø–æ –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–π —Å–ø–∏—Ä–∞–ª–∏, —Å–≤—è–∑–∞–Ω–Ω–æ–π —Å –§–∏–±–æ–Ω–∞—á—á–∏

üìç 5. –ò—Å–∫—É—Å—Å—Ç–≤–æ –∏ –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞
–ó–æ–ª–æ—Ç–æ–µ —Å–µ—á–µ–Ω–∏–µ (œÜ ‚âà 1.618) ‚Äî –ø—Ä–µ–¥–µ–ª –æ—Ç–Ω–æ—à–µ–Ω–∏—è —Å–æ—Å–µ–¥–Ω–∏—Ö —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏

–ü–∞—Ä—Ñ–µ–Ω–æ–Ω, –ú–æ–Ω–∞ –õ–∏–∑–∞, —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—ã–π –¥–∏–∑–∞–π–Ω ‚Äî –∏—Å–ø–æ–ª—å–∑—É—é—Ç –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ –§–∏–±–æ–Ω–∞—á—á–∏ –¥–ª—è –≥–∞—Ä–º–æ–Ω–∏–∏

–ò—Ç–æ–≥: –ó–∞—á–µ–º –≤–∞–º –∑–Ω–∞—Ç—å –§–∏–±–æ–Ω–∞—á—á–∏?
–û–±–ª–∞—Å—Ç—å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏—è	–ö–∞–∫ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è
–û–±—É—á–µ–Ω–∏–µ	–ë–∞–∑–∞ –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è —Ä–µ–∫—É—Ä—Å–∏–∏ –∏ –¥–∏–Ω–∞–º–∏—á–µ—Å–∫–æ–≥–æ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è
–ê–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö	–ü–æ–∏—Å–∫ —Ü–∏–∫–ª–∏—á–µ—Å–∫–∏—Ö –ø–∞—Ç—Ç–µ—Ä–Ω–æ–≤, –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–µ
–§–∏–Ω–∞–Ω—Å—ã	–¢–µ—Ö–Ω–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑, –ø–æ–∏—Å–∫ —É—Ä–æ–≤–Ω–µ–π –ø–æ–¥–¥–µ—Ä–∂–∫–∏
–ê—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –ü–û	–ü—Ä–∏–º–µ—Ä –¥–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
–ò–Ω—Ç–µ—Ä–≤—å—é	–ö–ª–∞—Å—Å–∏—á–µ—Å–∫–∞—è –∑–∞–¥–∞—á–∞ –Ω–∞ –∞–ª–≥–æ—Ä–∏—Ç–º—ã
–§–∏–±–æ–Ω–∞—á—á–∏ ‚Äî —ç—Ç–æ –Ω–µ –∞–±—Å—Ç—Ä–∞–∫—Ç–Ω–∞—è –º–∞—Ç–µ–º–∞—Ç–∏–∫–∞, –∞:
‚úÖ –ò–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –ø–∞—Ç—Ç–µ—Ä–Ω–æ–≤
‚úÖ –ü—Ä–∏–º–µ—Ä –¥–ª—è –∏–∑—É—á–µ–Ω–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤
‚úÖ –ú–æ—Å—Ç –º–µ–∂–¥—É –º–∞—Ç–µ–º–∞—Ç–∏–∫–æ–π –∏ —Ä–µ–∞–ª—å–Ω—ã–º –º–∏—Ä–æ–º
‚úÖ –¢–µ—Å—Ç –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è –æ—Å–Ω–æ–≤ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è

–ö–∞–∫ –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç—É-–∞–Ω–∞–ª–∏—Ç–∏–∫—É: –í—ã –±—É–¥–µ—Ç–µ –≤—Å—Ç—Ä–µ—á–∞—Ç—å –§–∏–±–æ–Ω–∞—á—á–∏ –ø—Ä–∏:

–ê–Ω–∞–ª–∏–∑–µ –≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —Ä—è–¥–æ–≤

–û–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π

–ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–∏ –ø—Ä–æ–≥–Ω–æ–∑–Ω—ã—Ö –º–æ–¥–µ–ª–µ–π

–û–±—Å—É–∂–¥–µ–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∏—á–µ—Å–∫–∏—Ö —Ä–µ—à–µ–Ω–∏–π –Ω–∞ —Å–æ–±–µ—Å–µ–¥–æ–≤–∞–Ω–∏—è—Ö

–≠—Ç–æ —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω–∞—è –∫–æ–Ω—Ü–µ–ø—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –ø–æ–º–æ–≥–∞–µ—Ç –¥—É–º–∞—Ç—å –∏ –æ —Ä–µ–∫—É—Ä—Å–∏–∏, –∏ –æ–± –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏, –∏ –æ –∫—Ä–∞—Å–æ—Ç–µ –º–∞—Ç–µ–º–∞—Ç–∏–∫–∏ –≤ –ø—Ä–∏—Ä–æ–¥–µ –∏ —Ç–µ—Ö–Ω–æ–ª–æ–≥–∏—è—Ö.
"""
```

```python
def fibonacci_iterative(n):
    """
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç n-–Ω–æ–µ —á–∏—Å–ª–æ –§–∏–±–æ–Ω–∞—á—á–∏, 
    –≥–¥–µ F(1) = 0, F(2) = 1.
    –ò—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –º–µ—Ç–æ–¥ (O(n)).
    """
    if n <= 0:
        raise ValueError("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º")
    if n == 1:
        return 0
    if n == 2:
        return 1
    
    prev_prev = 0  # F(1)
    prev = 1       # F(2)
    
    for i in range(3, n + 1):
        current = prev + prev_prev
        prev_prev, prev = prev, current
    
    return prev

# –†–µ–∫—É—Ä—Å–∏–≤–Ω–∞—è –≤–µ—Ä—Å–∏—è (–¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è)
def fibonacci_recursive(n):
    """
    –†–µ–∫—É—Ä—Å–∏–≤–Ω–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è (–º–µ–¥–ª–µ–Ω–Ω–æ –¥–ª—è –±–æ–ª—å—à–∏—Ö n).
    """
    if n <= 0:
        raise ValueError("n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º")
    if n == 1:
        return 0
    if n == 2:
        return 1
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è: –≤—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö k —á–∏—Å–µ–ª
def print_first_k_fibonacci(k):
    """
    –ü–µ—á–∞—Ç–∞–µ—Ç –ø–µ—Ä–≤—ã–µ k —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏ —Å –Ω—É–º–µ—Ä–∞—Ü–∏–µ–π —Å 1.
    """
    print(f"–ü–µ—Ä–≤—ã–µ {k} —á–∏—Å–µ–ª –§–∏–±–æ–Ω–∞—á—á–∏ (F(1)=0):")
    for i in range(1, k + 1):
        print(f"  F({i}) = {fibonacci_iterative(i)}")

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
if __name__ == "__main__":
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–µ—Ä–≤—ã—Ö 10 —á–∏—Å–µ–ª
    print_first_k_fibonacci(10)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞
    n = 4
    print(f"\nF({n}) = {fibonacci_iterative(n)}")  # –î–æ–ª–∂–Ω–æ –±—ã—Ç—å 2
    
    # –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω–æ–≥–æ –∏ —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ (–¥–ª—è –º–∞–ª—ã—Ö n)
    n_test = 8
    print(f"F({n_test}) –∏—Ç–µ—Ä–∞—Ç–∏–≤–Ω–æ: {fibonacci_iterative(n_test)}")
    print(f"F({n_test}) —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ: {fibonacci_recursive(n_test)}")
```
