```python
"""
–ß—Ç–æ —Ç–∞–∫–æ–µ –û –±–æ–ª—å—à–æ–µ? üéØ
O –±–æ–ª—å—à–æ–µ - —ç—Ç–æ —Å–ø–æ—Å–æ–± –æ–ø–∏—Å–∞—Ç—å, –∫–∞–∫ –±—ã—Å—Ç—Ä–æ —Ä–∞—Å—Ç—ë—Ç –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞ –∏–ª–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞–º—è—Ç–∏, 
–∫–æ–≥–¥–∞ –≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ —Å—Ç–∞–Ω–æ–≤—è—Ç—Å—è –ë–û–õ–¨–®–ï.

–ü—Ä–æ—â–µ –≥–æ–≤–æ—Ä—è: –ù–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –∞–ª–≥–æ—Ä–∏—Ç–º "—Ç–æ—Ä–º–æ–∑–Ω–µ—Ç", –∫–æ–≥–¥–∞ –¥–∞–Ω–Ω—ã—Ö —Å—Ç–∞–Ω–µ—Ç –≤ 10, 100 –∏–ª–∏ 1000 —Ä–∞–∑ –±–æ–ª—å—à–µ?

–ü—Ä–æ—Å—Ç–∞—è –∞–Ω–∞–ª–æ–≥–∏—è üì¶
–ü—Ä–µ–¥—Å—Ç–∞–≤—å, —á—Ç–æ —É —Ç–µ–±—è –µ—Å—Ç—å –∫–æ—Ä–æ–±–∫–∏, –∏ —Ç—ã –¥–æ–ª–∂–µ–Ω –∏—Ö –ø–µ—Ä–µ—Å—á–∏—Ç–∞—Ç—å:
"""
```

```python
# 1. O(1) ‚Äî "–ö–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–µ –≤—Ä–µ–º—è" (–ò–¥–µ–∞–ª—å–Ω–æ!)
def get_first_box(boxes):
    return boxes[0]  # –ë–µ—Ä—É –ø–µ—Ä–≤—É—é –∫–æ—Ä–æ–±–∫—É
```

```python
"""
–ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç: –°–∫–æ–ª—å–∫–æ –±—ã –∫–æ—Ä–æ–±–æ–∫ –Ω–∏ –±—ã–ª–æ (10, 1000, –º–∏–ª–ª–∏–æ–Ω) - —è –≤—Å–µ–≥–¥–∞ –¥–µ–ª–∞—é –æ–¥–Ω–æ –¥–µ–π—Å—Ç–≤–∏–µ.

–ê–Ω–∞–ª–æ–≥–∏—è: –ù–∞–π—Ç–∏ —Å–≤–æ–π –¥–æ–º –ø–æ –∞–¥—Ä–µ—Å—É. –î–æ–º–∞ 10 –∏–ª–∏ 10 –º–∏–ª–ª–∏–æ–Ω–æ–≤ - —è —Å—Ä–∞–∑—É –∏–¥—É –ø–æ –∏–∑–≤–µ—Å—Ç–Ω–æ–º—É –∞–¥—Ä–µ—Å—É.

–ì—Ä–∞—Ñ–∏–∫: –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–∞—è –ª–∏–Ω–∏—è. –í—Ä–µ–º—è –Ω–µ —Ä–∞—Å—Ç—ë—Ç –≤–æ–æ–±—â–µ!
"""
```

```python
# 2. O(n) ‚Äî "–õ–∏–Ω–µ–π–Ω–æ–µ –≤—Ä–µ–º—è" (–•–æ—Ä–æ—à–æ!)
def count_all_boxes(boxes):
    count = 0
    for box in boxes:  # –ü—Ä–æ—Ö–æ–∂—É –ø–æ –í–°–ï–ú –∫–æ—Ä–æ–±–∫–∞–º
        count += 1
    return count
```

```python
"""
–ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç: –ï—Å–ª–∏ –∫–æ—Ä–æ–±–æ–∫ –≤ 10 —Ä–∞–∑ –±–æ–ª—å—à–µ - –≤—Ä–µ–º–µ–Ω–∏ –Ω—É–∂–Ω–æ –≤ 10 —Ä–∞–∑ –±–æ–ª—å—à–µ.

–ê–Ω–∞–ª–æ–≥–∏—è: –ü—Ä–æ–≤–µ—Ä–∏—Ç—å –≤—Å–µ—Ö —É—á–µ–Ω–∏–∫–æ–≤ –≤ –∫–ª–∞—Å—Å–µ –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –¥–æ–º–∞—à–Ω–µ–π —Ä–∞–±–æ—Ç—ã. 
          10 —É—á–µ–Ω–∏–∫–æ–≤ ‚Äî 10 –ø—Ä–æ–≤–µ—Ä–æ–∫, 30 —É—á–µ–Ω–∏–∫–æ–≤ ‚Äî 30 –ø—Ä–æ–≤–µ—Ä–æ–∫.

–ì—Ä–∞—Ñ–∏–∫: –ü—Ä—è–º–∞—è –ª–∏–Ω–∏—è –ø–æ–¥ —É–≥–ª–æ–º 45
"""
```

```python
# 3. O(n¬≤) ‚Äî "–ö–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –≤—Ä–µ–º—è" (–ü–ª–æ—Ö–æ!)
def check_all_pairs(boxes):
    for box1 in boxes:        # –ü–µ—Ä–≤—ã–π —Ü–∏–∫–ª
        for box2 in boxes:    # –í—Ç–æ—Ä–æ–π —Ü–∏–∫–ª –≤–Ω—É—Ç—Ä–∏ –ø–µ—Ä–≤–æ–≥–æ!
            print(f"–ü–∞—Ä–∞: {box1} –∏ {box2}")
```

```python
"""
–ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç: –ï—Å–ª–∏ –∫–æ—Ä–æ–±–æ–∫ –≤ 10 —Ä–∞–∑ –±–æ–ª—å—à–µ - –≤—Ä–µ–º–µ–Ω–∏ –Ω—É–∂–Ω–æ –≤ 100 —Ä–∞–∑ –±–æ–ª—å—à–µ! (10¬≤ = 100)

–ê–Ω–∞–ª–æ–≥–∏—è: –£—Å—Ç—Ä–æ–∏—Ç—å —Ç—É—Ä–Ω–∏—Ä –ø–æ —Ä—É–∫–æ–ø–æ–∂–∞—Ç–∏—è–º –≤ –∫–ª–∞—Å—Å–µ. –ö–∞–∂–¥—ã–π –¥–æ–ª–∂–µ–Ω –ø–æ–∂–∞—Ç—å —Ä—É–∫—É –∫–∞–∂–¥–æ–º—É. 10 —á–µ–ª–æ–≤–µ–∫ ‚Äî 45 —Ä—É–∫–æ–ø–æ–∂–∞—Ç–∏–π. 20 —á–µ–ª–æ–≤–µ–∫ ‚Äî —É–∂–µ 190 —Ä—É–∫–æ–ø–æ–∂–∞—Ç–∏–π!

–ì—Ä–∞—Ñ–∏–∫: –ö—Ä—É—Ç–∞—è –ø–∞—Ä–∞–±–æ–ª–∞ (—Ä–∞—Å—Ç—ë—Ç –æ—á–µ–Ω—å –±—ã—Å—Ç—Ä–æ)
"""
```

```python
"""
–ü—Ä–∏–º–µ—Ä –∏–∑ –∂–∏–∑–Ω–∏ üçï
–ó–∞–¥–∞—á–∞: –ù–∞–π—Ç–∏ –¥—Ä—É–≥–∞ –≤ —Ç–æ–ª–ø–µ.

O(n) ‚Äî "–õ–∏–Ω–µ–π–Ω—ã–π –ø–æ–∏—Å–∫":

–°–º–æ—Ç—Ä—é –Ω–∞ –∫–∞–∂–¥–æ–≥–æ —á–µ–ª–æ–≤–µ–∫–∞ –ø–æ –æ—á–µ—Ä–µ–¥–∏

10 —á–µ–ª–æ–≤–µ–∫ ‚Üí –ø–æ—Å–º–æ—Ç—Ä—é –Ω–∞ 10 –ª–∏—Ü

100 —á–µ–ª–æ–≤–µ–∫ ‚Üí –ø–æ—Å–º–æ—Ç—Ä—é –Ω–∞ 100 –ª–∏—Ü

–í 10 —Ä–∞–∑ –±–æ–ª—å—à–µ –ª—é–¥–µ–π = –≤ 10 —Ä–∞–∑ –±–æ–ª—å—à–µ –≤—Ä–µ–º–µ–Ω–∏


O(log n) ‚Äî "–ë–∏–Ω–∞—Ä–Ω—ã–π –ø–æ–∏—Å–∫" (–û—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ!):

–ü—Ä–µ–¥—Å—Ç–∞–≤—å, —á—Ç–æ –ª—é–¥–∏ —Å—Ç–æ—è—Ç –≤ –∞–ª—Ñ–∞–≤–∏—Ç–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ

–°–º–æ—Ç—Ä—é –≤ —Å–µ—Ä–µ–¥–∏–Ω—É, –ø–æ–Ω–∏–º–∞—é, —á—Ç–æ –º–æ–π –¥—Ä—É–≥ "–ò–≤–∞–Ω–æ–≤" –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å–ø—Ä–∞–≤–∞ (–æ—Ç "–ö")

–û—Ç–±—Ä–∞—Å—ã–≤–∞—é –ª–µ–≤—É—é –ø–æ–ª–æ–≤–∏–Ω—É, —Å–º–æ—Ç—Ä—é –≤ —Å–µ—Ä–µ–¥–∏–Ω—É –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è

100 —á–µ–ª–æ–≤–µ–∫ ‚Üí –Ω—É–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –ø—Ä–∏–º–µ—Ä–Ω–æ 7 –ª–∏—Ü (–ø–æ—Ç–æ–º—É —á—Ç–æ 2‚Å∑ ‚âà 100)

1000 —á–µ–ª–æ–≤–µ–∫ ‚Üí –Ω—É–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –ø—Ä–∏–º–µ—Ä–Ω–æ 10 –ª–∏—Ü

–í 10 —Ä–∞–∑ –±–æ–ª—å—à–µ –ª—é–¥–µ–π = –Ω—É–∂–Ω–æ –≤—Å–µ–≥–æ –Ω–∞ ~3 –ø—Ä–æ–≤–µ—Ä–∫–∏ –±–æ–ª—å—à–µ!
"""
```

```python
# –ê —Ç–µ–ø–µ—Ä—å –ø—Ä–æ —Ñ—É–Ω–∫—Ü–∏—é –§–∏–±–æ–Ω–∞—á—á–∏ üìâ

def fib(n):
    if n <= 1:
        return n
    return fib(n-1) + fib(n-2)  # O(2^n) - –ö–ê–¢–ê–°–¢–†–û–§–ê!
# O(2‚Åø) ‚Äî "–≠–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è" (–ö–∞—Ç–∞—Å—Ç—Ä–æ—Ñ–∞!)
```

```python
"""
–ß—Ç–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç:

fib(5) –¥–µ–ª–∞–µ—Ç ~30 –≤—ã–∑–æ–≤–æ–≤

fib(10) –¥–µ–ª–∞–µ—Ç ~180 –≤—ã–∑–æ–≤–æ–≤

fib(20) –¥–µ–ª–∞–µ—Ç ~22,000 –≤—ã–∑–æ–≤–æ–≤

fib(30) –¥–µ–ª–∞–µ—Ç ~2.7 –ú–ò–õ–õ–ò–û–ù–ê –≤—ã–∑–æ–≤–æ–≤!

fib(40) –¥–µ–ª–∞–µ—Ç ~330 –ú–ò–õ–õ–ò–û–ù–û–í –≤—ã–∑–æ–≤–æ–≤!

–ê–Ω–∞–ª–æ–≥–∏—è: –°–ª—É—Ö–∏ –≤ —à–∫–æ–ª–µ. –ö–∞–∂–¥—ã–π, –∫—Ç–æ —É–∑–Ω–∞–ª —Å–µ–∫—Ä–µ—Ç, —Ä–∞—Å—Å–∫–∞–∑—ã–≤–∞–µ—Ç –µ–≥–æ –î–í–£–ú –¥—Ä—É–≥–∏–º. 
–ß–µ—Ä–µ–∑ —á–∞—Å –∑–Ω–∞—é—Ç 2 —á–µ–ª–æ–≤–µ–∫–∞, —á–µ—Ä–µ–∑ 2 —á–∞—Å–∞ ‚Äî 4, —á–µ—Ä–µ–∑ 3 —á–∞—Å–∞ ‚Äî 8... 
–ß–µ—Ä–µ–∑ 10 —á–∞—Å–æ–≤ ‚Äî —É–∂–µ 1024 —á–µ–ª–æ–≤–µ–∫–∞!

–ü–æ—á–µ–º—É —Ç–∞–∫ –ø–ª–æ—Ö–æ:

fib(100) –ø–æ—Ç—Ä–µ–±–æ–≤–∞–ª–æ –±—ã –±–æ–ª—å—à–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π, —á–µ–º –∞—Ç–æ–º–æ–≤ –≤–æ –í—Å–µ–ª–µ–Ω–Ω–æ–π

–ö–æ–º–ø—å—é—Ç–µ—Ä –Ω–µ –¥–æ–∂–∏–ª –±—ã –¥–æ –∫–æ–Ω—Ü–∞ –≤—ã—á–∏—Å–ª–µ–Ω–∏–π

–®–∫–∞–ª–∞ —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ –æ—Ç –ª—É—á—à–µ–≥–æ –∫ —Ö—É–¥—à–µ–º—É üìä
text
–ë—ã—Å—Ç—Ä–æ ‚Üí –ú–µ–¥–ª–µ–Ω–Ω–æ
O(1)      ‚Üê –ò–¥–µ–∞–ª—å–Ω–æ! (–∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–µ)
O(log n)  ‚Üê –û—Ç–ª–∏—á–Ω–æ! (–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–µ)
O(n)      ‚Üê –•–æ—Ä–æ—à–æ (–ª–∏–Ω–µ–π–Ω–æ–µ)
O(n log n) ‚Üê –ù–æ—Ä–º–∞–ª—å–Ω–æ (–ª–∏–Ω–µ–π–Ω–æ-–ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫–æ–µ)
O(n¬≤)     ‚Üê –ü–ª–æ—Ö–æ (–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ)
O(2‚Åø)     ‚Üê –£–∂–∞—Å–Ω–æ (—ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ)
O(n!)     ‚Üê –ö–∞—Ç–∞—Å—Ç—Ä–æ—Ñ–∞ (—Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª—å–Ω–æ–µ)
"""
```
