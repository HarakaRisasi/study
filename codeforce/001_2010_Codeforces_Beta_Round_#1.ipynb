**Ограничение по времени на тест:** 2 seconds  
**Ограничение по памяти на тест:** 64 megabytes  
**Ввод:** standard input  
**Вывод:** standard output  

Театральная площадь в столице Берляндии представляет собой прямоугольник  
_n×m_ метров. По случаю очередного юбилея города было принято решение о  
замощении площади квадратными гранитными плитами. Каждая плита имеет  
размер _a×a_.  

Какое наименьшее количество плит понадобится для замощения площади?  

Разрешено покрыть плитами большую поверхность, чем театральная площадь,  
но она должна быть покрыта обязательно. Гранитные плиты нельзя ломать или  
дробить, а разрешено использовать только целиком. Границы плит должны  
быть параллельны границам площади.  

**Входные данные**  
В первой строке записано три целых натуральных числа _n_, _m_ и _a_  
(1 ≤ _n_, _m_, _a_ ≤ 10^9).  

**Выходные данные**  
Выведите искомое количество плит.  

**Примеры тестов**  

**входные данные**  
6 6 4  

**выходные данные**  
4  

## Задача

Нужно покрыть прямоугольник **n × m** квадратными плитами **a × a**.  
Плиты кладутся без наложения, параллельно сторонам, ломать их нельзя.

---

### Шаг 1. Покрытие по длине

Вдоль стороны **n** поместится:

⌈n / a⌉ плит.

---

### Шаг 2. Покрытие по ширине

Вдоль стороны **m** поместится:

⌈m / a⌉ плит.

---

### Шаг 3. Общее количество плит

answer = ⌈n / a⌉ × ⌈m / a⌉

---

### Шаг 4. Проверка на примере

Пример: n = 6, m = 6, a = 4

⌈6 / 4⌉ = ⌈1.5⌉ = 2  
⌈6 / 4⌉ = 2  
answer = 2 × 2 = 4

Совпадает с примером.

---

### Шаг 5. Реализация

Поскольку n, m, a могут быть до 10⁹, прямое вычисление в целых числах  
с использованием целочисленного деления — самый быстрый способ.

Формула для целых чисел:

⌈n / a⌉ = (n + a - 1) // a   (целочисленное деление)

Тогда:

answer = ((n + a - 1) // a) × ((m + a - 1) // a)

```python
n, m, a = map(int, input().split())

# Вычисляем количество плит по длине и ширине
count_n = (n + a - 1) // a
count_m = (m + a - 1) // a


# Общее количество плит
result = count_n * count_m

print(result)
```
